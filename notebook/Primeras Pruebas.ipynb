```python
import pandas as pd

df = pd.read_excel("../spreadsheets/jordan_review_for_e_rav.xlsx")

# Saco las columnas que no voy a usar
cols_to_drop = ['site_id', 'MPN (ignore)']
df = df.drop(cols_to_drop, axis="columns")

# Agrego el https:// a todos los links
df['product_line_url'] = df['product_line_url'].apply(lambda s: 'https://'+s)
df['product_url'] = df['product_url'].apply(lambda s: 'https://'+s)
df
```

```python
try: 
    from BeautifulSoup import BeautifulSoup
except ImportError:
    from bs4 import BeautifulSoup

import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry

parsed_html, html = "", ""
# Recorro todas las filas y voy buscando las URLs
for index, row in df.iterrows():
    #print(index, row['product_url'])
    # Obtengo el HTML del producto (product_url)
    r = requests.get(row['product_url'])
    html = r.text
    
    parsed_html = BeautifulSoup(html)
    b = parsed_html.body
    
    match = b.find(id='w1-5-_msg')
    
    if match is not None:     
        print(index, row['product_url'], match)
    
    
    if index > 10000:
        break
    
```
